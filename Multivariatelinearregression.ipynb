```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices (1).csv')
df
```

```python
# I wanted to replace null values with median val 
# if there is decimal values import math and use math.floor
import math
df.bedrooms.median()
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df['price']) #features = x (so that should be in 2D [[]] and y will be in series [])
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
reg.predict(pd.DataFrame({
    'area': [3000],
    'bedrooms': [3],
    'age': [20]
}))
```

```python
from word2number import w2n
```

```python
fd.rename(columns = {
    'test_score(out of 10)' : 't_score',
    'interview_score(out of 10)' : 'i_score'
}, inplace = True )
fd.head()
```

```python
fd.experience = fd.experience.fillna('zero')
fd
```

```python
import math 

test_median = math.floor(fd.t_score.median())
print(test_median)
interview_median = math.floor(fd.i_score.median())
print(interview_median)
```

```python
fd.t_score = fd.t_score.fillna(test_median) #i am converting null values to median score 
fd
```

```python
fd.i_score = fd.i_score.fillna(interview_median)
fd
```

```python
#It is telling that i need to change experience columns from to string to int val 
fd.experience= fd.experience.astype(str)
fd.experience = fd.experience.apply(w2n.word_to_num)
fd
```

```python
#Now that i have all filled null val, i need to train the model using sklear and fit
regg = linear_model.LinearRegression()
regg.fit(fd[['experience', 't_score', 'i_score']], fd['salary($)'])
```

```python
regg.predict(pd.DataFrame({
    'experience' : [2],
    't_score' : [9],
    'i_score' : [6]
}))
# Now that i have predicted salary for 2yr exp, 9testscore, 6interview score
```
